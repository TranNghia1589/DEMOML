## 1. Xử lý null

```python
# Thay dấu '?' bằng NaN để xử lý giá trị thiếu
import numpy as np

df.replace('?', np.nan, inplace=True)

# Kiểm tra số lượng giá trị thiếu trong mỗi cột
df.isnull().sum()
```

```python
# Hiển thị các cột có giá trị null
df.columns[df.isnull().any()]
```

```python
# Hiển thị các dòng có giá trị null
df[pd.isnull(df).any(axis=1)]
```

```python
df['workclass'].value_counts()
```

```python
df['workclass'].isnull().sum()
```

```python
# Thay những giá trị null của cột 'workclass' thành mode của workclass
df['workclass'] = df['workclass'].fillna(df['workclass'].mode()[0])
```

```python
df['occupation'].value_counts()
```

```python
# Thay những giá trị null của cột 'occupation' thành mode của occupation
df['occupation'] = df['occupation'].fillna(df['occupation'].mode()[0])
```

```python
df['native.country'].value_counts()
```

```python
# Thay những giá trị null của cột 'native.country' thành mode của native.country
df['native.country'] = df['native.country'].fillna(df['native.country'].mode()[0])
```

```python
# Xây dựng các nhóm phân loại
education_map = {
    'Preschool': 'Basic Education',
    '1st-4th': 'Basic Education',
    '5th-6th': 'Basic Education',
    '7th-8th': 'Basic Education',
    '9th': 'Basic Education',
    '10th': 'Basic Education',
    '11th': 'Basic Education',
    '12th': 'Basic Education',
    'HS-grad': 'High School',
    'Some-college': 'Post-secondary Education',
    'Assoc-voc': 'Post-secondary Education',
    'Assoc-acdm': 'Post-secondary Education',
    'Bachelors': 'Undergraduate',
    'Masters': 'Graduate Education',
    'Doctorate': 'Graduate Education'
}

# Áp dụng phân loại vào cột 'education'
df['education_grouped'] = df['education'].map(education_map)
```
